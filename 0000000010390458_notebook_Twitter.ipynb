```python
import pandas as pd
import numpy as np
```

```python
twitter_data_train=pd.read_csv('training_twitter_x_y_train.csv')
twitter_data_test=pd.read_csv('test_twitter_x_test.csv')
```

```python
print(twitter_data_train.head(15))
print(twitter_data_test.head(5))
```

```python
tweet_text_train=np.array(twitter_data_train['text'])
tweet_text_test=np.array(twitter_data_test['text'])
```

```python
y_train=np.array(twitter_data_train['airline_sentiment'])
```

```python
tweet_count_train=np.array(twitter_data_train['retweet_count'])
tweet_count_test=np.array(twitter_data_test['retweet_count'])
```

```python
x_train=[]
x_test=[]
for i in range(len(tweet_count_train)):
    x_train.append((tweet_text_train[i]+' ')*(1))
for i in range(len(tweet_count_test)):
    x_test.append((tweet_text_test[i]+' ')*(1))
```

```python
from nltk.corpus import stopwords,wordnet
from string import punctuation
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
stop=set(stopwords.words('english'))
punctuations=punctuation
stop.update(punctuations)
```

```python
from nltk import pos_tag
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    if tag.startswith('N'):
        return wordnet.NOUN
    if tag.startswith('R'):
        return wordnet.ADV
    if tag.startswith('V'):
        return wordnet.VERB
    return wordnet.NOUN
```

```python
def clean_tweets(x):
    output_words=[]
    for w in x.split():
        if w.lower() not in stop:
            pos=pos_tag([w])
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_word)
    return output_words
```

```python
x_train_cleaned = [clean_tweets(x) for x in x_train]
x_test_cleaned = [clean_tweets(x) for x in x_test]
```

```python
print(x_train_cleaned[8],'\n',x_train[8])
print(x_test_cleaned[8],']n',x_test[8])
```

```python
x_train=[' '.join(tweets) for tweets in x_train_cleaned]
x_test=[' '.join(tweets) for tweets in x_test_cleaned]
```

```python
print(x_train[11],y_train[11])
print(x_test[8])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vec=CountVectorizer(max_features=2000)
print(x_train[0],y_train[0])
```

```python
x_train_features=count_vec.fit_transform(x_train)
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
x_test_features=count_vec.transform(x_test)
```

```python
x_test_features.todense()
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(C=5)
svc.fit(x_train_features,y_train)
```

```python
y_pred=svc.predict(x_test_features)
print(y_pred)
```

```python
np.savetxt('outputSVC.csv',y_pred,'%s')
```
